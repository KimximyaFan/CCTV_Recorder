```python
import cv2 as cv
import numpy as np

video_file = 'rtsp://210.99.70.120:1935/live/cctv003.stream'
video = cv.VideoCapture(video_file)

if video.isOpened() == False:
    exit()

width = int(video.get(3))
height = int(video.get(4))

isRecord = False
img = None    
fps = 50
wait_msec = int( 1000 / fps )
path = 'temp.avi'
fourcc = cv.VideoWriter_fourcc('X', 'V', 'I', 'D')
out = cv.VideoWriter(path, fourcc, fps, ( width, height ) )

contrast = 1.6
contrast_step = 0.1
brightness =-40
brightness_step = 1


# ===================================================================

def Record(frame):
    out.write(img)
    center = ( int (8 * width / 9), int(height / 7) )
    cv.circle(frame, center, radius=20, color=(0, 0, 255), thickness=-1)
    cv.imshow('Video Player', frame)

def Intesity_Transform( frame ):
    img_tran = contrast * frame + brightness
    img_tran[img_tran < 0] = 0
    img_tran[img_tran > 255] = 255
    img_tran = img_tran.astype(np.uint8)
    return img_tran
    
while True:
    valid, img = video.read()
    key = cv.waitKey(wait_msec)
    
    if not valid:
        break
        
    if key == 27:
        break
    elif key == ord(' '):
        isRecord = not isRecord
    elif key == ord('+') or key == ord('='):
        contrast += contrast_step
    elif key == ord('-') or key == ord('_'):
        contrast -= contrast_step
    elif key == ord(']') or key == ord('}'):
        brightness += brightness_step
    elif key == ord('[') or key == ord('{'):
        brightness -= brightness_step
    elif key == ord('>') or key == ord('.'):
        fps = fps + 1
    elif key == ord('<') or key == ord(','):
        fps = fps - 1
        
    img = Intesity_Transform(img)
    
    cv.imshow('Video Player', img)
    
    if isRecord == True:
        Record(img)
        
video.release()
out.release()
cv.destroyAllWindows()
```
